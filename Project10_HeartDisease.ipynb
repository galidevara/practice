```python
from scipy.stats import zscore
dt=abs(zscore(data))
print(data.shape)
data_new=data.loc[(dt<3).all(axis=1)]
data_new.shape
```

```python
data_new
```

```python
import numpy as np
for i in data_new.columns:
    if data_new.skew().loc[i]>0.55:
        data_new[i]=np.sqrt(data_new[i])
```

```python
for i in data_new.columns:
    if data_new.skew().loc[i]>0.55:
        data_new[i]=np.log(data_new[i])
```

```python
data_new.skew()
```

```python
x=data_new.drop(columns=['fbs','exang','ca','num'])
y=data_new[['num']]
y=y.astype(int)
print(x.shape)
print(y.shape)
```

```python

from sklearn.metrics import r2_score,accuracy_score,confusion_matrix,classification_report
from sklearn.model_selection import cross_val_score,train_test_split
def max_r2score(reg,x,y):
    maxr2=0
    for r_state in range(45,100):
        x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=r_state,test_size=0.20)
        reg.fit(x_train,y_train)
        y_pred=reg.predict(x_test)
        r2scr=r2_score(y_test,y_pred)
        if r2scr>maxr2:
            maxr2=r2scr
            final=r_state
            print('for the randon state:',final,'max r2scr is',maxr2)
    print(cross_val_score(reg,x,y,cv=5).mean())
    print(lg.score(x_train,y_train))
    print(accuracy_score(y_test,y_pred))
    print(confusion_matrix(y_test,y_pred))
    print(classification_report(y_test,y_pred))
```

```python
from sklearn.linear_model import LogisticRegression
lg=LogisticRegression()
max_r2score(lg,x,y)
```

```python
from sklearn.neighbors import KNeighborsClassifier
kn=KNeighborsClassifier()
max_r2score(kn,x,y)
```

```python
from sklearn.tree import DecisionTreeClassifier
dtc=DecisionTreeClassifier()
max_r2score(dtc,x,y)
```

```python
from sklearn.naive_bayes import GaussianNB
gnb=GaussianNB()
max_r2score(gnb,x,y)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import GradientBoostingRegressor
grb=GradientBoostingRegressor()
parameters={'learning_rate':[0.001,0.01,0.1,1],'n_estimators':[50,500,1000,1500]}
clf=GridSearchCV(grb,parameters,cv=10)
clf.fit(x,y)
clf.best_params_
```

```python
maxr2=0
for r_state in range(45,100):
    x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=r_state)
    grb=GradientBoostingRegressor(learning_rate=0.001,n_estimators=1500)
    grb.fit(x_train,y_train)
    y_pred=grb.predict(x_test)
    r2scr=r2_score(y_test,y_pred)
    if r2scr>maxr2:
        maxr2=r2scr
        final=r_state
        print('for the randon state:',final,'max r2scr is',maxr2)
print(cross_val_score(grb,x,y,cv=10).mean())
```
