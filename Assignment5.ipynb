```python
import string
import matplotlib.pyplot as plt
from collections import Counter
```

```python
def get_tweets():
    import GetOldTweets3 as got
    tweetcriteria=got.manager.TweetCriteria().setQuerySearch("#Coronavirus")\
        .setSince("2020-01-01")\
        .setUntil("2020-01-31")\
        .setMaxTweets(150)
    tweets=got.manager.TweetManager.getTweets(tweetcriteria)
    
    text_tweets=[[tweet.text] for tweet in tweets]
    
    return text_tweets
```

```python
#!pip install GetOldTweets3
```

```python
text=""
tweets=get_tweets()

lenth=len(tweets)
for t in range(0,lenth):
    text=tweets[t][0]+" "+text
    
print(text)
```

```python
lower=text.lower()
filtertext= lower.translate(str.maketrans(" "," ",string.punctuation))
print(filtertext)
```

```python
import nltk
tokenize=nltk.word_tokenize(filtertext)
```

```python
from nltk.corpus import stopwords
stop_words=stopwords.words('english')

final_words=[word for word in tokenize if word not in stop_words]
final_words
```

```python
emotion_list=[]

with open ('emotions.txt','r') as file:
    for line in file:
        em=line.replace("\n","").replace(",","").replace("'","").strip()
        word, emotion=em.split(':')
        if word in final_words:
            emotion_list.append(emotion)
            
w = Counter(emotion_list)
print(w)
```

```python
fig, ax1 = plt.subplots()
ax1.bar(w.keys(), w.values())
fig.autofmt_xdate()
plt.show()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
import pandas as pd
```

```python
def bow(strng, vect):
    string=vect.fit_transform(strng)
    return pd.DataFrame(string.toarray(),columns=vect.get_feature_names())
```
