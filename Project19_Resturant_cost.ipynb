```python
dt['RATING']=dt['RATING'].str.replace('NEW', '1')
dt['RATING']=dt['RATING'].str.replace('-', '1').astype(float)
```

```python
dt['VOTES']=dt['VOTES'].str.replace(' votes','').astype(float)
```

```python
dt['CITY'].fillna('missing',inplace=True)  
dt['RATING'].fillna(3.5, inplace=True)  
dt['VOTES'].fillna(0.0, inplace=True)
```

```python
cuisines_list=[]
for i in dt['CUISINES']:
    cuisines_list.append(list(i.split(',')))

dt['CUISINES']=cuisines_list
```

```python
dt_cuisines=dt['CUISINES'].apply(lambda x: pd.Series(1, x))
```

```python
dt_cuisines
```

```python
title_list=[]
for i in dt['TITLE']:
    title_list.append(list(i.split(',')))
    
dt['TITLE']=title_list
dt_title=dt['TITLE'].apply(lambda x: pd.Series(1, x))
dt_title
```

```python
dt=pd.concat([dt,dt_cuisines,dt_title], axis=1)
```

```python
dt.drop(columns=['CUISINES','LOCALITY','TIME','TITLE','RESTAURANT_ID'],inplace=True)
```

```python
dt.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
dt['CITY']=le.fit_transform(dt['CITY'])

dt['CITY'].head()
```

```python
dt.isnull().sum()
```

```python
train_f = dt[dt['COST'].isnull()!=True]
test_f = dt[dt['COST'].isnull()==True]
test_f.drop('COST', axis=1, inplace=True)
```

```python
test_f.shape
```

```python
x=train_f.drop(columns='COST')
y=train_f['COST']
```

```python
x.fillna(0,inplace=True)
x.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split,cross_val_score
from sklearn.metrics import r2_score,mean_squared_error
import math
```

```python
def max_r2(reg,x,y):
    maxsrc=0
    for r in range(45,100):
        x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=r)
        reg.fit(x_train,y_train)
        y_pred=reg.predict(x_test)
        r2src=r2_score(y_test,y_pred)
        if maxsrc<r2src:
            maxsrc=r2src
            final=r
            print("for final state",final,'max r2 score is:',maxsrc)    
    print('Cross val score:',cross_val_score(reg,x,y,cv=10).mean())
    
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
max_r2(lr,x,y)
```

```python
from sklearn.tree import DecisionTreeRegressor
dct=DecisionTreeRegressor()
max_r2(dct,x,y)
```

```python
from sklearn.neighbors import KNeighborsRegressor
kn=KNeighborsRegressor()
max_r2(kn,x,y)
```

```python
from sklearn.ensemble import RandomForestRegressor
rfr=RandomForestRegressor()
max_r2(rfr,x,y)
```

```python
from sklearn.ensemble import GradientBoostingRegressor
gbr=GradientBoostingRegressor()
max_r2(gbr,x,y)
```
