```python
x=data_new.drop(columns=['value'])
y=data_new['value']
```

```python
from sklearn.model_selection import train_test_split,cross_val_score
from sklearn.metrics import r2_score,accuracy_score,confusion_matrix,classification_report
```

```python
def max_r2(reg,x,y):
    maxsrc=0
    for r in range(45,100):
        x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=r)
        reg.fit(x_train,y_train)
        y_pred=reg.predict(x_test)
        r2src=r2_score(y_test,y_pred)
        if maxsrc<r2src:
            maxsrc=r2src
            final=r
            print("for final state",final,'max r2 score is:',maxsrc)    
    print('score of model:',reg.score(x_train,y_train))
    print('Cross val score:',cross_val_score(reg,x,y,cv=5).mean())
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
max_r2(lr,x,y)
```

```python
from sklearn.tree import DecisionTreeRegressor
dct=DecisionTreeRegressor()
max_r2(dct,x,y)
```

```python
from sklearn.neighbors import KNeighborsRegressor
kn=KNeighborsRegressor()
max_r2(kn,x,y)
```

```python
from sklearn.ensemble import RandomForestRegressor
rfr=RandomForestRegressor()
max_r2(rfr,x,y)
```
