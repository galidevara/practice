```python
from scipy.stats import zscore
dt=abs(zscore(data))
print(dt.shape)
data_new=data.loc[(dt<3).all(axis=1)]
data_new.shape
```

```python
data_new.skew()
```

```python
import numpy as np
for i in data_new.columns:
    if data_new.skew().loc[i]>0.55:
        data_new[i]=np.log(data_new[i])
        
```

```python
x=data_new.drop(columns=['Proline'])
y=data_new[['Proline']]
print(x.shape)
print(y.shape)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.neighbors import KNeighborsRegressor
import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn.metrics import mean_squared_error,r2_score
max_score=0

for r in range (45,100):
    x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=r,test_size=0.20)
    lr=LinearRegression()
    lr.fit(x_train,y_train)
    y_pred=lr.predict(x_test)
    r2scr=r2_score(y_test,y_pred)
    if r2scr>max_score:
        max_score=r2scr
        final_state=r
print(lr.score(x_train,y_train))  
print("max_r2score for",final_state,'is',max_score)

from sklearn.model_selection import cross_val_score
print(cross_val_score(lr,x,y,cv=5).mean())
```

```python
from sklearn.metrics import mean_squared_error,r2_score
max_score=0

for r in range (45,100):
    x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=r,test_size=0.20)
    dct=DecisionTreeRegressor()
    dct.fit(x_train,y_train)
    y_pred=dct.predict(x_test)
    r2scr=r2_score(y_test,y_pred)
    if r2scr>max_score:
        max_score=r2scr
        final_state=r
print(dct.score(x_train,y_train))  
print("max_r2score for",final_state,'is',max_score)

from sklearn.model_selection import cross_val_score
print(cross_val_score(dct,x,y,cv=10).mean())
```

```python
max_score=0

for r in range (45,100):
    x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=r,test_size=0.20)
    kn=KNeighborsRegressor()
    kn.fit(x_train,y_train)
    y_pred=kn.predict(x_test)
    r2scr=r2_score(y_test,y_pred)
    if r2scr>max_score:
        max_score=r2scr
        final_state=r
print(kn.score(x_train,y_train))  
print("max_r2score for",final_state,'is',max_score)

from sklearn.model_selection import cross_val_score
print(cross_val_score(kn,x,y,cv=10).mean())
```

```python
from sklearn.ensemble import AdaBoostRegressor
from sklearn.model_selection import GridSearchCV
adb=AdaBoostRegressor()
parameters={'learning_rate':[0.001,0.01,0.1,1],'n_estimators':[10,100,500,1000]}
clf=GridSearchCV(adb,parameters,cv=10)
clf.fit(x,y)
clf.best_params_
```

```python
adb=AdaBoostRegressor(learning_rate=0.1,n_estimators=1000)
print('cors_val_score_r2 mean:',cross_val_score(adb,x,y,cv=5,scoring='r2').mean())
print('cors_val_score std:',cross_val_score(adb,x,y,cv=5).std())
```

```python
def maxscore(reg,x,y):
    max_score=0
    for r in range (45,100):
        x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=r,test_size=0.20)
        reg.fit(x_train,y_train)
        y_pred=reg.predict(x_test)
        r2scr=r2_score(y_test,y_pred)
        if r2scr>max_score:
            max_score=r2scr
            final_state=r
print("max_r2score for",final_state,'is',max_score)
```

```python
r=maxscore(adb,x,y)
```

```python
from sklearn.ensemble import GradientBoostingRegressor
gbr=GradientBoostingRegressor()
parameters={'learning_rate':[0.001,0.01,0.1,1],'n_estimators':[10,100,500,1000]}
clf=GridSearchCV(gbr,parameters,cv=10)
clf.fit(x,y)
clf.best_params_
```

```python
gbr=GradientBoostingRegressor(learning_rate=0.1,n_estimators=500)
print('cors_val_score mean:',cross_val_score(gbr,x,y,cv=5,scoring='r2').mean())
print('cors_val_score std:',cross_val_score(gbr,x,y,cv=5).std())
```

```python
r=maxscore(gbr,x,y)
```
