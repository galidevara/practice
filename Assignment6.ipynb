```python
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE

words=['intelligence','characteristic','generalize','programmed','proficiency','speed','tasks','diagnosis','engines',
       'medical','simplest','digger','food','often','trait','chiefly','number','might','recall','tense']

def splot(model):
    labels=[]
    wordvec=[]
    for word in words:
        wordvec.append(model[word])
        labels.append(word)
        
    tsne_mod=TSNE(perplexity=3,n_components=2,init='pca',random_state=42)
    coord=tsne_mod.fit_transform(wordvec)
        
    x=[]
    y=[]
        
    for value in coord:
        x.append(value[0])
        y.append(value[1])
            
    plt.figure(figsize=(15,8)) 
    for i in range(len(x)):
        plt.scatter(x[i],y[i])
        plt.annotate(labels[i],
                     xy=(x[i], y[i]),
                     xytext=(2, 2),
                     textcoords='offset points', ha='left')

    plt.show()

splot(model)
```

```python
text = [[word for word in doc.lower().split() if word not in str]
         for document in corp]
```

```python
from collections import defaultdict

freq = defaultdict(int)
freq
```

```python
import pprint
for t in text:
    for token in t:
        freq[token] += 1
        
filterd = [[token for token in t if freq[token] > 1] for t in text]
pprint.pprint(filterd)
```
