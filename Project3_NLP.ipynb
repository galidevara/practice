```python
url1="https://economictimes.indiatimes.com/news/defence/united-states-to-slap-sanctions-on-over-two-dozen-targets-tied-to-iran-arms/articleshow/78222036.cms"
response=requests.get(url1)
soup1=BeautifulSoup(response.text,'html.parser')
content1=soup1.find_all('div',{"data-brcount":"67"})
```

```python
content1
```

```python
train=''
for i in content1:
    train=train+' '+i.text
    
train
```

```python
import nltk
train=nltk.sent_tokenize(train)
```

```python
train[:2]
```

```python
import gensim
from gensim.utils import simple_preprocess
from gensim.parsing.preprocessing import STOPWORDS
from nltk.stem import WordNetLemmatizer,SnowballStemmer
```

```python
def preprocess(text):
    result=[]
    for token in gensim.utils.simple_preprocess(text):
        if token not in gensim.parsing.preprocessing.STOPWORDS and len(token)>3:
            result.append(lematize(token))
            
    return result

stemmer=SnowballStemmer('english')
def lematize(text):
    return stemmer.stem(WordNetLemmatizer().lemmatize(text,pos='v'))
```

```python
processed_doc=[]

for doc in train:
    processed_doc.append(preprocess(doc))
```

```python
processed_doc[:4]
```

```python
dictionary=gensim.corpora.Dictionary(processed_doc)
```

```python
print(dictionary)
```

```python
#dictionary.filter_extremes(no_below=15,no_above=0.1,keep_tokens=10000)
```

```python
bagofcorpus=[dictionary.doc2bow(doc) for doc in processed_doc]
```

```python
bagofcorpus[2:4]
```

```python
lda_model=gensim.models.LdaMulticore(bagofcorpus,num_topics=10,id2word=dictionary,passes=5,
                                    workers=2)
```

```python
for idx,topic in lda_model.print_topics():
    print('Topic : {} \nWords : {}'.format(idx,topic))
    print('\n')
```

```python
bow_vector = dictionary.doc2bow(preprocess(test))
```

```python
for index, score in sorted(lda_model[bow_vector]):
    print("Score: {}\t Topic: {}".format(score, lda_model.print_topic(index, 30)))
```
