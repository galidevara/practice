```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
crime['DISTRICT']=le.fit_transform(crime['DISTRICT'])
crime['STATE/UT']=le.fit_transform(crime['STATE/UT'])
```

```python
crime.skew()
```

```python
from scipy.stats import zscore
dt=abs(zscore(crime))
print(crime.shape)
crime_new=crime.loc[(dt<3).all(axis=1)]
print(crime_new.shape)
```

```python
crime_new.skew()
```

```python
for i in crime_new.columns:
    if crime_new.skew().loc[i]>0.55:
        crime_new[i]=np.sqrt(crime_new[i])
        
print(crime_new.skew())
```

```python
x=crime_new.drop(columns=['TOTAL IPC CRIMES'])
y=crime_new['TOTAL IPC CRIMES']
```

```python
from sklearn.model_selection import train_test_split,cross_val_score
from sklearn.metrics import r2_score
```

```python
def max_r2(reg,x,y):
    maxsrc=0
    for r in range(45,100):
        x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=r)
        reg.fit(x_train,y_train)
        ypred=reg.predict(x_test)
        r2src=r2_score(y_test,ypred)
        if maxsrc<r2src:
            maxsrc=r2src
            final=r
            print("for final state",final,'max r2 score is:',maxsrc)
    print('score of the model:',reg.score(x_train,y_train))

    print('Cross val score:',cross_val_score(reg,x,y,cv=10).mean())
  
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
max_r2(lr,x,y)
```

```python
from sklearn.tree import DecisionTreeRegressor
dct=DecisionTreeRegressor()
max_r2(dct,x,y)
```

```python
from sklearn.neighbors import KNeighborsRegressor
kn=KNeighborsRegressor()
max_r2(kn,x,y)
```

```python
from sklearn.ensemble import RandomForestRegressor
rfr=RandomForestRegressor()
max_r2(rfr,x,y)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
from sklearn.ensemble import GradientBoostingRegressor
grb=GradientBoostingRegressor()
parameters={'learning_rate':[0.001,0.01,0.1,1],'n_estimators':[10,50,100,1000]}
clf=GridSearchCV(grb,parameters,cv=10)
clf.fit(x,y)
clf.best_params_
```

```python
grb=GradientBoostingRegressor(learning_rate=0.1,n_estimators=1000)
max_r2(grb,x,y)
```
